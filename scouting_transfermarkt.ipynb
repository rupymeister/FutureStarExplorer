```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
import time
import numpy as np
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

csv_path = "fifa_23_players_data.csv"
df = pd.read_csv(csv_path)



```

```python
from selenium.common.exceptions import NoSuchElementException


known_as = df['Known As'].to_numpy().astype(str)
full_name = df['Full Name'].to_numpy().astype(str)
full_names_lowercase = np.char.lower(full_name)
known_as_lowercase = np.char.lower(known_as)
filtered_player_names = known_as[np.char.find(known_as, '.') == -1]
filtered_player_names_lowercase = np.char.lower(filtered_player_names)

print(filtered_player_names_lowercase)
print(full_names_lowercase)
```

```python

driver = webdriver.Chrome()
prices = np.array([])
for i in range(0,len(full_names_lowercase)-1):
    url = "https://www.transfermarkt.com.tr/schnellsuche/ergebnis/schnellsuche?query=" + full_names_lowercase[i].replace(" ", "+")
    driver.get(url)

    try:
        price = driver.find_element(By.XPATH, '/html/body/div[2]/main/div[1]/div/div/div[1]/div/table/tbody/tr/td[6]')
        prices = np.append(prices, price.text)
        print(full_names_lowercase[i], ": ", price.text)
    except NoSuchElementException:
        url = "https://www.transfermarkt.com.tr/schnellsuche/ergebnis/schnellsuche?query=" + known_as_lowercase[i].replace(" ", "+")
        driver.get(url)
        try:   
            price = driver.find_element(By.XPATH, '/html/body/div[2]/main/div[1]/div/div/div[1]/div/table/tbody/tr/td[6]')
            prices = np.append(prices, price.text)
            print(known_as_lowercase[i], ": ", price.text)
            
        except NoSuchElementException:
            prices = np.append(prices, "-")
            print(known_as_lowercase[i], ": Skipped")
            
    if(i % 500 == 0):
        df['Market Value'] = pd.Series(prices[:len(df)]).astype(str)
        df.loc[(df['Market Value'] == 'nan') | (df['Market Value'] == 'Skipped'), 'Market Value'] = '-'
        df.to_csv('transfermarkt_1.csv', index = False)
        
        
    
    
```
